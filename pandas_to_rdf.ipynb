```python
from rdflib import Graph, Literal, RDF, URIRef
# rdflib knows about some namespaces, like FOAF
from rdflib.namespace import FOAF , XSD

# create a Graph
gl = Graph()

# Create an RDF URI node to use as the subject for multiple triples
donna = URIRef("http://example.org/donna")

# Add triples using store's add() method.
gl.add((donna, RDF.type, FOAF.Person))
gl.add((donna, FOAF.nick, Literal("donna", lang="ed")))
gl.add((donna, FOAF.name, Literal("Donna Fales")))
gl.add((donna, FOAF.mbox, URIRef("mailto:donna@example.org")))

# Add another person
ed = URIRef("http://example.org/edward")

# Add triples using store's add() method.
gl.add((ed, RDF.type, FOAF.Person))
gl.add((ed, FOAF.nick, Literal("ed", datatype=XSD.string)))
gl.add((ed, FOAF.name, Literal("Edward Scissorhands")))
gl.add((ed, FOAF.mbox, URIRef("mailto:e.scissorhands@example.org")))

# Add another person
h1 = URIRef("http://example.org/harits")

# Add triples using store's add() method.
gl.add((h1, RDF.type, FOAF.Person))
gl.add((h1, FOAF.nick, Literal("H1", datatype=XSD.string)))
gl.add((h1, FOAF.name, Literal("Harits Abdr", 'Ais')))
gl.add((h1, FOAF.mbox, URIRef("mailto:harits@example.org")))
```

```python
# Iterate over triples in store and print them out.
print("--- printing raw triples ---")
for s, p, o in gl:
    # print((s, p, o))
    print(f'{s} <-> {p} <-> {o}')
```

```python
s.n3
```

```python
# For each foaf:Person in the store, print out their mbox property's value.
print("--- printing mboxes ---")
for person in gl.subjects(RDF.type, FOAF.Person):
    for mbox in gl.objects(person, FOAF.mbox):
        print(person, mbox)

# Bind the FOAF namespace to a prefix for more readable output
gl.bind("foaf", FOAF)
```

```python
# print all the data in the Notation3 format
print("--- printing mboxes ---")
print(gl.serialize(format='turtle').decode("utf-8"))
```
